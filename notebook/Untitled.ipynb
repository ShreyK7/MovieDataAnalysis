```python
from sqlalchemy import create_engine
from sqlalchemy import text

conn_string = "mysql+pymysql://{user}:{password}@{host}/".format(
    host="db.ipeirotis.org", user="student", password="dwdstudent2015"
)

engine = create_engine(conn_string)
```

```python
#create a new table to store our informations.
db_name = "public"
table_name = "ag8298_oscar_general"
# Create a table
create_table_query = f"""CREATE TABLE IF NOT EXISTS {db_name}.{table_name}
                                (
                                  year VARCHAR(50),
                                  year_awarded VARCHAR(50),
                                  category VARCHAR(50),
                                  name VARCHAR(50),
                                  film VARCHAR(50),
                                  winner VARCHAR(50)
                                  ) """

with engine.connect() as connection:
  connection.execute(text(create_table_query))
```

```python
# LOAD THE DATASET 
import numpy as np
```

```python
df = pd.read_csv("data/the_oscar_award.csv")
df.replace(to_replace = np.nan, value = " ")  
```

```python
for ind in df.index:
    
    print(df['year_film'][ind]," , ", df['year_ceremony'][ind]," , ", df['category'][ind]," , ", df['name'][ind]," , ", df['film'][ind]," , ", df['winner'][ind] )

    
```

```python
#inserting into the database now.

query_template = f"""
                    INSERT IGNORE INTO
                    {db_name}.{table_name}(year, year_awarded, category, name, film, winner)
                    VALUES (:year, :year_awarded, :category, :name, :film, :winner)
                  """
#NOTE TO SELF : the ':' after each values is actually a placeholder to indicate values will be dynamically filled afterwards.

with engine.connect() as connection:
  for ind in df.index:
      query_parameters = {
          "year": df['year_film'][ind] if pd.notna(df['year_film'][ind]) else 'None',
          "year_awarded": df['year_ceremony'][ind] if pd.notna(df['year_ceremony'][ind]) else 'None',
          "category": df['category'][ind] if pd.notna(df['category'][ind]) else 'None',
          "name": df['name'][ind] if pd.notna(df['name'][ind]) else 'None',
          "film": df['film'][ind] if pd.notna(df['film'][ind]) else 'None',
          "winner": df['winner'][ind] if pd.notna(df['winner'][ind]) else 'None'
      }
      print(df['year_film'][ind]," , ", df['year_ceremony'][ind]," , ", df['category'][ind]," , ", df['name'][ind]," , ", df['film'][ind]," , ", df['winner'][ind])      
      connection.execute(text(query_template), query_parameters)
  connection.commit()
```
